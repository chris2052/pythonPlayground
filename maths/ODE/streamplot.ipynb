```python
from matplotlib.pyplot import cm
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [9,6]
plt.rcParams['figure.dpi'] = 300 # 200 e.g. is really fine, but slower
```

```python
nx, ny = .3, .3
x = np.arange(-3, 3, nx)
y = np.arange(-2, 2, ny)
X, Y = np.meshgrid(x, y)
dy = X + np.sin(Y)
dx = np.ones(dy.shape)

color = dy
lw = 1

plt.streamplot(X,Y,dx, dy, color=color, density=1., cmap='jet', arrowsize=1)
```

```python
plt.quiver(X, Y, dx, dy, color='Teal', headlength=7)
```
