# Phase Space Visualization

```python
from pylab import *
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,6]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

## First example

```python
xvalues, yvalues = meshgrid(arange(0, 3, 0.1), arange(0, 3, 0.1))
```

$$
\begin{align}
\dfrac{dx}{dt} &= x - xy \\
\dfrac{dy}{dt} &= -y + xy
\end{align}
$$

```python
xdot = xvalues - xvalues*yvalues
ydot = -yvalues + xvalues*yvalues
```

`streamplot` is preciesely designed for dwawing **phase space** of continous-time models

```python
streamplot(xvalues, yvalues, xdot, ydot)
```

## Motion of simple pendulum
$$
\dfrac{d^2\theta}{dt^2} = -\dfrac{g}{L}\sin(\theta)
$$

```python
g = -9.81
L = 2
theta_values, theta_dot_values = meshgrid(arange(-8, 8, 0.1), arange(-3, 3, 0.1))
```

```python
theta_dot = theta_dot_values
theta_ddot = -g/L*sin(theta_values)
```

```python
# fig, ax = plt.subplots(0, 0)
# ax[0, 0].axis('equal')
```

```python
# ax[1,1].streamplot(theta_values, theta_dot_values, theta_dot, theta_ddot)
```

```python
streamplot(theta_values, theta_dot_values, theta_dot, theta_ddot)
```
