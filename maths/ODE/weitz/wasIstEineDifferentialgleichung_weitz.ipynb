```python
from math import exp
from plot import *
```

```python
TU = 20
c = -0.1
d = 10
```

```python
def Temp (T0):
    return lambda t: (T0 - TU)*exp(c*t) + TU
```

```python
# sofortige Zuckerzugabe
f = Temp(95 - d)
```

```python
# ohne Zucker
g1 = Temp(95)
```

```python
g2 = Temp(g1(2) - d)
```

```python
g = lambda t: g1(t) if t<= 2 else g2(t-2)
```

```python
plotFunc2D([f, g, lambda t: 65], [0, 5], samples = 100)
```
