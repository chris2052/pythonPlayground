# Generate slope fields in Python

$y' = \dfrac{x}{y}$  
or  
$y' = x + y$

```python
import numpy as np
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = [9,6]
plt.rcParams['figure.dpi'] = 200 # 200 e.g. is really fine, but slower
```

```python
# Differential equation
# diff = y'= y/x (or say x+y)
def diff(x,y):
    return y+x # try also x+y

x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
```

```python
# use x,y
for j in x:
    for k in y:
        slope = diff(j,k)
        domain = np.linspace(j-0.07,j+0.07,2)
        def fun(x1,y1):
            z = slope*(domain-x1)+y1
            return z
        plt.plot(domain,fun(j,k),solid_capstyle='projecting',solid_joinstyle='bevel')

plt.title("Slope field y'")
plt.grid(True)
plt.show()
```
